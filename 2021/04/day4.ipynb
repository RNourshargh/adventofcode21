```python
## pt1 e.g.
file = "eg.txt"
calls = []
with open(file) as f:
    first_line = f.readline()
    for i in first_line.strip().split(','):
        calls.append(int(i))
    # print(calls)

a = np.loadtxt(file, dtype = int, skiprows = 1)
n = np.shape(a)[0]//5
boards = np.reshape(a, (n,5,5))
rows = np.zeros((n,5), dtype = int)
cols = np.zeros((n,5), dtype = int)
rolling_sum = np.zeros((n), dtype = int)

for call in calls:
        locs = np.nonzero(boards == call)
        rolling_sum[locs[0]] += call
        rows[locs[0],locs[1]] +=1
        cols[locs[0], locs[2]] +=1
        if np.amax(rows) == 5:
            winner =np.unravel_index(rows.argmax(), rows.shape)[0]
            break
        if np.amax(cols) == 5:
            winner =np.unravel_index(cols.argmax(), cols.shape)[0]
            break

board_score = boards[winner].sum()-rolling_sum[winner]
print("board score", board_score)
print("last call", call)
print("score", board_score * call)
```
